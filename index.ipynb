## Exercise 6

Let's consider the example where 2 dice are thrown. Given that **at least one** of the dice has come up on a number higher than 4, what is the probability that the sum is 8?

Let i,j be the numbers shown on the dice . Denote events as below:

* **Event A is when Either i Or j is 5 or 6** (Keep an eye on either - or)
* **Event B is when i + j = 8**


* What is the size of sample space Ω ?
* What is P(A ∩ B) ?
* What is P(A) ?
* Use above to calculate P(B|A) .

```python
# Your solution

# size of sample space?
sample_space = 6**2

# Event A - where d1 or d2 is 5 or 6
event_A = [(1,5), (2,5), (3,5), (4,5), (5,5), 
          (5,1), (5,2), (5,3), (5,4), (5,5),
          (1,6), (2,6), (3,6), (4,6), (6,6),
          (6,1), (6,2), (6,3), (6,4), (6,6)]
len(event_A)/sample_space

# Event B size
event_B = [(5,3), (6,2), (3,5), (6,2)]
len(event_B)

# P(A and B)
p_A_and_B = len(event_B) / sample_space

# P(A) = 1/3
p_A = len(event_A)/sample_space

# P(B given A) = P(A and B) / P(A)
p_B_given_A = p_A_and_B / p_A

print("Sample space = ", sample_space)
print("Event A = ", len(event_A))
print("Event B = ", len(event_B))
print("Prob A & B = ", p_A_and_B)
print("Prob A = ", p_A)
print("Prob B given A", p_B_given_A)
```

## Exercise 7

Let's consider a credit card example. At a supermarket, we randomly select customers and make notes of whether a given customer owns a Visa card (event A) or an Amex credit card (event B). Some customers own both cards.
You can assume that:

- P(A) = 0.5
- P(B) = 0.4
- both A and B = 0.25.


With the knowledge we have about conditional probabilities, compute and interpret the following probabilities:

- $P(B \mid A)$
- $P(B'\mid A)$
- $P(A\mid B)$
- $P(A'\mid B)$


```python
# Your solution
# P(A and B) = P(A) * P(B given A)
p_A = 0.5
p_B = 0.4
p_notA = 1-p_A
p_notB = 1-p_B
p_A_and_B = 0.25
p_A_and_notB = p_A*p_notB
p_B_and_notA = p_B*p_notA

p_B_given_A = p_A_and_B / p_A
p_A_given_B = p_A_and_B / p_B

p_notB_given_A = 1-p_B_given_A # CORRECT ANSWER: 1-p_B_given_A = 0.5 (i.e. complement of p_B_given_A)
p_notA_given_B = 1-p_A_given_B #CORRECT ANSWER: 1-p_A_given_B = 0.375

print("p_B_given_A (correct)", p_B_given_A)
print("p_A_given_B (correct)", p_A_given_B)
print("p_notB_given_A", p_notB_given_A)
print("p_notA_given_B", p_notA_given_B)
```
